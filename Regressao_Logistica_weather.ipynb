```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('weather.csv')
```

```python
df
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df['Sunshine'].fillna(df['Sunshine'].mean(),inplace=True)
```

```python
df['WindGustSpeed'].fillna(df['WindGustSpeed'].mean(),inplace=True)
```

```python
df['WindSpeed9am'].fillna(df['WindSpeed9am'].mean(),inplace=True)
```

```python
df.drop(df[['WindGustDir','WindDir9am','WindDir3pm']],axis=1,inplace=True)
```

```python
df['RainToday'].replace('No',0,inplace=True)
```

```python
df['RainToday'].replace('Yes',1,inplace=True)
```

```python
df['RainTomorrow'].replace('No',0,inplace=True)
```

```python
df['RainTomorrow'].replace('Yes',1,inplace=True)
```

```python
df
```

```python
sns.jointplot(x='Evaporation',y='MaxTemp',data=df)
```

```python
sns.jointplot(x='Evaporation',y='Humidity3pm',data=df,kind='kde',color='red')
```

```python
sns.pairplot(df, hue='MaxTemp')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['MaxTemp','WindSpeed3pm','Evaporation','Humidity3pm','Cloud3pm','Humidity3pm','Pressure3pm']]
y=df['RainToday']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predict = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,predict))
print(confusion_matrix(y_test,predict))
```
